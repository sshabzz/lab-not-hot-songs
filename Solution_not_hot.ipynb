```python
# Extracting artists names in a list
artist = []
for tag in range(0, len(soup.select("td > a")),4):
    artist.append(soup.select("td > a")[tag].get_text())

    
# artist
print("Total number of artists:", len(artist))
```

```python
# Here I am storing links in a list

links = []
for i in range (1,16,1):
    start=str(i)
    if i == 1: 
         url = "http://www.acclaimedmusic.net/061024/1948-02s.htm"
    else:
        url = "http://www.acclaimedmusic.net/061024/1948-02s" + start + ".htm"
        
    links.append(url)

links


```

```python
# Extracting artist & songs from the given links
songs = []
artists = []

for url in links:
    response = requests.get(url)
    if ( response.status_code == 200 ):
        # parse all pages
        soup = BeautifulSoup(response.content, "html.parser")
        for i in range(0,len(soup.select("td a")),4):
            artists.append(soup.select("td a")[i].get_text())
        for i in range(1,len(soup.select("td a")),4):
            songs.append(soup.select("td a")[i].get_text())
        
print(len(songs))   
print(len(artists)) 
```

```python
# Creating a dataframe for not_hot songs:

not_hot = pd.DataFrame({'Title':songs,'Artist':artists})
display(not_hot.shape)
display(not_hot.isna().sum())
not_hot.head()
```

```python
# Exporting CSV - Save as CSV

not_hot.to_csv('not_hot.csv',index=False)
```
